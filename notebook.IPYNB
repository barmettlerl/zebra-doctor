```python
import csv
import pandas as pd
from cycler import cycler
import matplotlib.pyplot as plt
import matplotlib as mpl
# add marker cycler to the default prop_cycle
mpl.rcParams.update(mpl.rcParamsDefault)
custom_cycler = (cycler(color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']) + cycler(linestyle=['-', '--', ':', '-.', '-']))

import scienceplots
plt.style.use(['science','ieee'])
%matplotlib inline
```

## Simple test

```python
simple_no_backup = pd.read_csv('simple_results/simple_no_backup.csv')
simple_no_backup_cpu_stats_df = pd.read_csv('simple_results/simple_no_backup_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_no_backup['cpu'] = simple_no_backup_cpu_stats_df

simple_no_backup_dashmap = pd.read_csv('simple_results/simple_no_backup_dashmap.csv')
simple_no_backup_dashmap_cpu_stats_df = pd.read_csv('simple_results/simple_no_backup_dashmap_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_no_backup_dashmap['cpu'] = simple_no_backup_dashmap_cpu_stats_df

simple_file_backup = pd.read_csv('simple_results/simple_file_backup.csv')
simple_file_backup_cpu_stats_df = pd.read_csv('simple_results/simple_file_backup_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_file_backup['cpu'] = simple_file_backup_cpu_stats_df

simple_pickledb = pd.read_csv('simple_results/simple_pickledb.csv')
simple_pickledb_cpu_stats_df = pd.read_csv('simple_results/simple_pickledb_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_pickledb['cpu'] = simple_pickledb_cpu_stats_df

simple_rocksdb_wal = pd.read_csv('simple_results/simple_rocksdb_wal.csv')
simple_rocksdb_wal_cpu_stats_df = pd.read_csv('simple_results/simple_rocksdb_wal_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_rocksdb_wal['cpu'] = simple_rocksdb_wal_cpu_stats_df

simple_single_rocksdb = pd.read_csv('simple_results/simple_single_rocksdb.csv')
simple_single_rocksdb_cpu_stats_df = pd.read_csv('simple_results/simple_single_rocksdb_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_single_rocksdb['cpu'] = simple_single_rocksdb_cpu_stats_df

simple_okaywal = pd.read_csv('simple_results/simple_okaywal.csv')
simple_okaywal_cpu_stats_df = pd.read_csv('simple_results/simple_okaywal_cpu_stats.csv').mean(axis=1).mean(axis=0)
simple_okaywal['cpu'] = simple_okaywal_cpu_stats_df

simple_results = pd.concat([simple_no_backup, simple_no_backup_dashmap, simple_file_backup, simple_pickledb, simple_rocksdb_wal, simple_single_rocksdb, simple_okaywal])

# add name column
simple_results['name'] = ['simple_no_backup',  'simple_no_backup_dashmap' , 'simple_file_backup', 'simple_pickledb', 'simple_rocksdb_wal', 'simple_single_rocksdb', 'simple_okaywal']

# make duration in seconds
simple_results['duration'] = simple_results['duration'] / 1000

simple_results
```
